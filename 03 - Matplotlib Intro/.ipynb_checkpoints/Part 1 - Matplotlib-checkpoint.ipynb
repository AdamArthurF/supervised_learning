```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
x
```

```python
y
```

# Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y1)
ax.plot(x, y2)
```

### shortcut

```python
plt.figure(figsize=(8,8))
plt.plot(x, y1, 'r-.')
plt.plot(x, y2, 'b.')
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.axis('equal')
plt.title('Sin dan Cos')
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes()

ax.plot(x, y1, 'r-.')
ax.plot(x, y2, 'b.')
ax.set_xlim(0, 10)
ax.set_ylim(-2, 2)
ax.set_title("sin dan cos")
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes()

ax.plot(x, y1, 'r-.')
ax.plot(x, y2, 'b.')
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos", xlabel = 'x', ylabel = 'y')

ax.axis('equal')
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes()

ax.plot(x, y1, 'r-.', label='sin')
ax.plot(x, y2, 'b.', label='cos')
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos", xlabel = 'x', ylabel = 'y')

ax.axis('equal')
```

## Scatter Plot

```python
df = pd.read_csv("data/circle_data.csv", delimiter=";")
```

```python
df.head()
```

```python
x = df.x
y = df.y
rad = df.radius
```

```python
plt.scatter(x, y)
plt.axis('equal')
```

```python
plt.scatter(x, y, c='r', s=10)
plt.axis('equal')
```

```python
plt.scatter(x, y, c=rad, s=rad*90)
plt.axis('equal')
```

# Histogram

```python
numbers = np.random.rand(10000)
```

```python
plt.hist(numbers, bins=25)
```

```python
numbers = np.random.randn(10000)
plt.hist(numbers, bins=25)
```

```python
numbers = np.random.randn(100000)
plt.hist(numbers, bins=80, density=1)
```

```python
numbers = np.random.randn(100000)
plt.hist(numbers, bins=80, density=1, cumulative=True)
```

# Subplot

## MATLAB Style

```python
plt.figure(figsize=(15, 10))

plt.subplot(231)
plt.hist(numbers, bins=80, density=1)

plt.subplot(233)
plt.hist(numbers, bins=80, density=1)

plt.subplot(235)
plt.scatter(x, y, c=rad, s=rad*90)
plt.axis('equal')

plt.subplot(236)
plt.hist(numbers, bins=80, density=1)
```

## Matplotlib Style

```python
fig, ax = plt.subplots(2,3, figsize=(15,10))

ax[0,0].hist(numbers, bins=80, density=1)
ax[0,1].hist(numbers, bins=80, density=1)
ax[1,1].scatter(x, y, c=rad, s=rad*90)
ax[1,2].hist(numbers, bins=80, density=1)
```
