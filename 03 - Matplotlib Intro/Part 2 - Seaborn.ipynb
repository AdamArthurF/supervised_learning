```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/linear.csv")
df.head()
```

```python
df1 = df[df.tipe == 'data1']
plt.scatter(df1.X, df1.y)

df2 = df[df.tipe == 'data2']
plt.scatter(df2.X, df2.y)
```

# Seaborn

```python
plt.figure(figsize=(6,6))
sns.scatterplot("X", "y", data=df, hue="tipe")
```

```python
sns.lmplot('X', 'y', data=df, hue="tipe")
```

```python
sns.distplot(df1.y, bins=20)
```

# Seaborn & Pandas Dataframe

## Statistical Summary

```python
df = pd.read_csv('data/cereal.csv', index_col='name')
df.head()
```

## Missing value

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.isna(), cmap='Blues')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), cmap='bwr', vmin=-1,vmax=1, annot=True, fmt=".1f")
```
