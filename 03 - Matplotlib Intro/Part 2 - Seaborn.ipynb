# Seaborn & Pandas Dataframe

## Statistical Summary

```python
df = pd.read_csv('data/cereal.csv', index_col='name')
df.head()
```

## Missing value

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.isna(), cmap='Blues')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), cmap='bwr', vmin=-1,vmax=1, annot=True, fmt=".1f")
```

## Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='shelf', vars=['calories', 'fat', 'rating'])
```

```python
sns.pairplot(df, hue='shelf', vars=['calories', 'fat', 'rating'], kind='reg')
```

```python
sns.jointplot('potass', 'fiber', data=df)
```

## Categorical Data

### Persebaran 

```python
sns.boxplot('mfr', 'rating', data=df)
```

## Count

```python
sns.catplot('mfr', data=df, kind='count')
```

```python
sns.countplot('mfr', data=df, kind='shelf')
```
