```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
```

# Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y1)
ax.plot(x, y2)
```

### shortcut

```python
plt.figure(figsize=(8,8))
plt.plot(x, y1, 'r-.')
plt.plot(x, y2, 'b.')
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.axis('equal')
plt.title('Sin dan Cos')
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes()

ax.plot(x, y1, 'r-.')
ax.plot(x, y2, 'b.')
ax.set_xlim(0, 10)
ax.set_ylim(-2, 2)
ax.set_title("sin dan cos")
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes()

ax.plot(x, y1, 'r-.')
ax.plot(x, y2, 'b.')
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos", xlabel = 'x', ylabel = 'y')

ax.axis('equal')
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes()

ax.plot(x, y1, 'r-.', label='sin')
ax.plot(x, y2, 'b.', label='cos')
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos", xlabel = 'x', ylabel = 'y')

ax.axis('equal')
```
